```python
import pandas as pd
import matplotlib.pyplot as plt
import pymongo
%matplotlib inline
```

```python
client = pymongo.MongoClient("localhost")
db = client["lianjia"]
fangyuan = db["fangyuanbigsz"]
```

```python
data = pd.DataFrame(list(fangyuan.find()))
```

```python
data.columns
```

```python
data.shape
```

```python
df1 = data.loc[:,["成交时间","成交均价","行政区域"]]
```

```python
df1.head()
```

```python
df1["行政区域"].value_counts()
```

```python
from pyecharts import Bar
attr = df1["行政区域"].value_counts().index.tolist()
v1 = df1["行政区域"].value_counts().values.tolist()
bar = Bar("深圳各区域房屋成交数量柱状图")
bar.add("深圳房价", attr, v1, is_stack=True)
bar.render("深圳各区域成交数量.html")
bar
```

```python
longgang = df1[df1["行政区域"]=="龙岗区"]
futian = df1[df1["行政区域"]=="福田区"]
nanshan = df1[df1["行政区域"]=="南山区"]
baoan = df1[df1["行政区域"]=="宝安区"]
luohu = df1[df1["行政区域"]=="罗湖区"]
longhua = df1[df1["行政区域"]=="龙华区"]
yantian = df1[df1["行政区域"]=="盐田区"]
```

```python
longgang.head()
```

```python
longgang1 = longgang.sort_values(by="成交时间")
longgang1["成交均价"] = longgang1["成交均价"].astype("int32")
longgang1 = longgang1.groupby("成交时间").mean()
longgang1.index = pd.to_datetime(longgang1.index)
longgang1 = longgang1.resample("M").mean()
longgang1["成交均价"] = longgang1["成交均价"].apply(lambda x : int(x))
```
