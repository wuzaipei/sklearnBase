```python
import pandas as pd
import matplotlib.pyplot as plt
import pymongo
%matplotlib inline
```

```python
client = pymongo.MongoClient("localhost")
db = client["lianjia"]
fangyuan = db["fangyuanbigsz"]
```

```python
data = pd.DataFrame(list(fangyuan.find()))
```

```python
data.columns
```

```python
data.sort_values(by="成交时间",inplace=True)
```

```python
df1 = data.loc[:,["成交时间","成交均价","小区"]]
df1["成交均价"] = df1["成交均价"].astype("int32")
```

```python
df1
```

```python
xiaoqu = df1.groupby("小区")#.agg(lambda x:x.max()-x.min())
df2 = xiaoqu["成交均价"].agg(lambda x:(x.max()-x.min())/x.min())
```

```python
df3 = df2.sort_values(ascending=False)
```

```python
df3 = df3.apply(lambda x:"%.2f" % x)
```

```python
from pyecharts import Bar
attr = df3.index.tolist()
v = df3.values.tolist()
bar = Bar("深圳各小区房屋成交价格排行图")
bar.add("深圳房价", attr, v, is_stack=True)
bar.render("深圳各小区成交价格排行.html")
bar
```
